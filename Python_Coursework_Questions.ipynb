### Question 3

Use three different methods to print the numbers 200 to 300.

````python
# Method 1 - for loop
print("Method 1")
for n in range(200, 301):
    print(n)

# Method 2 - while loop
print("\nMethod 2")
num = 200
while num <= 300:
    print(num)
    num += 1

# Method 3 - Recursion
print("\nMethod 3")

def number_printer(start, stop):
    """
    Print the numbers between the first argument start and the last argument stop.
    """
    print(start)
    if start == stop: # Escape case to stop at the provided stop argument
        return 
    else: # Rerun the function with the next value
        number_printer(start + 1, stop)

# Use my made function to print numbers between 200 and 300
number_printer(200, 300)
````

### Question 6
Using the startwars dataset imported below, identify the individuals with tallest height and lowest mass.

````python
import pandas as pd
starwars_dataset="https://www.fabricionarcizo.com/post/starwars/updated_starwars.csv"
starwars = pd.read_csv(starwars_dataset)
````

````python
# Subset the data frame to select the name and height columns.
starwars_nh = starwars[["name", "height"]]

# Identify the tallest individual
max_height = starwars_nh["height"].max()
tallest_individual = starwars_nh[starwars_nh["height"] == max_height] # Obtain the row where we only have the max height. This variable is a dataframe.
tall_name = tallest_individual["name"].iloc[0] # From the name column of the data frame, obtain the first name (string).

print(f"The tallest individual is {tall_name} with a height of {max_height} cm. \n")

# Identify the shortest individual
min_height = starwars_nh["height"].min()
shortest_individual = starwars_nh[starwars_nh["height"] == min_height] # Obtain the row where we only have the min height.
short_name = shortest_individual["name"].iloc[0] # From the name column of the data frame, obtain the first name (string).

print(f"The shortest individual is {short_name} with a height of {min_height} cm.")
````

### Question 7
Analyze the life expectancy dataset (which can be loaded by running the cell below).

Calculate the number of unique countries in the dataset.

Calculate the country and year with the highest life expectancy.

Calculate the year with the highest total expenditure.

Provide your code using pandas.


````python
import seaborn as sns
import pandas as pd
life_expectancy = sns.load_dataset('healthexp')
````

````python
# Calculate the number of uniques countries in the dataset
unique_countries = life_expectancy["Country"].unique() # Obtain list of unique countries in the dataframe

print(f"There are {len(unique_countries)} unique countries in this dataframe.")


# Calculate the country and year with the highest life expectancy.
max_life_expectancy = life_expectancy["Life_Expectancy"].max() # Obtain the max life expectancy value
max_le_row = life_expectancy[life_expectancy["Life_Expectancy"] == max_life_expectancy] # Select the row with the max life expectancy. Returns a dataframe.
max_le_country = max_le_row["Country"].iloc[0] # From the max_le_row dataframe, obtain the country.
max_le_year = max_le_row["Year"].iloc[0] # From the max_le_row dataframe, obtain the year.

print(f"The country with the highest life expectancy of {max_life_expectancy}, was {max_le_country} during {max_le_year}.")


# Calculate the year with the highest total expendature
annual_sum = life_expectancy.groupby("Year")["Spending_USD"].sum() # Group by each year and find the sum of the spending for all countries combined.
sorted_annual_sums = annual_sum.sort_values(ascending=False) # Sort the annual sums by largest first
max_annual_sum = sorted_annual_sums.iloc[0] # obtain max value
year_of_max_spend = sorted_annual_sums.index[0] # obtain year

print(f"The highest total expenditure of ${round(max_annual_sum, 2)} occurred during {year_of_max_spend}.")
````

### Question 10
Write a function that will convert carat to grams and find the heaviest diamond.

1 carat = 0.2 grams

````python
def carat_converter(carat):
    """
    Converts carats into grams, where 1 carat = 0.2g.
    """
    return(carat * 0.2)


# Find the heaviest diamond
diamonds["weight"] = diamonds["carat"].apply(carat_converter) # Add new column to the dataframe and apply the carat_converter function to all the carat values to give the weight
max_weight = diamonds["weight"].max()

print(diamonds) # Check weight column is present
print(f"\nThe heaviest diamond weighs {max_weight}g.")
````
